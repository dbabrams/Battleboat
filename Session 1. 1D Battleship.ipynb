# 1D Battleship

```python
# enter an opponent guess
# Reminder: add code to prevent invalid guesses (guesses that fall outside of our grid) from the opponent
op_guess = input("Guess the location of the ship: ")
```

```python
# create a list for the first row (Row A) of our game of battleship with 6 columns
rowA = [0,0,1,1,0,0]
```

```python
# translating "Battleship" numbering system, which is 1-indexed, to "Python" 0-indexing
col = int(op_guess[-1])-1

# determine if the result was a hit or a miss
if rowA[col] == 1:
    print(op_guess + " is a hit")
else:
    print(op_guess + " is a miss")
```
